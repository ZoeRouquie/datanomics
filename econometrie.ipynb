```python
from random import uniform
import matplotlib.pyplot as plt

v = 100
t = 10
c = 10
m = 1
n = 2
benchmark = c + t 


def tarification(p_concu):
    return (p_concu+t+c)/2
    
```

Sc√©nario 1:Equilibre de NASH

```python
p_a =  uniform(c,v)
p_b = uniform(c,v)
i_tild = (p_b-p_a+t)/(2+t)
prix_a = [p_a]
prix_b = [p_b]
demandes_a = [i_tild]
demandes_b = [(1-i_tild)]
profits_a = [(p_a-c)*i_tild]
profits_b = [(p_b-c)*(1-i_tild)]


for _ in range(30):
    p_a = tarification(p_b)
    p_b = tarification(p_a)
    i_tild = (p_b-p_a+t)/(2+t)
    prix_a.append(p_a)
    prix_b.append(p_b)
    demandes_a.append(i_tild)
    demandes_b.append(1-i_tild)
    profits_a.append((p_a-c)*i_tild)
    profits_b.append((p_b - c)*(1-i_tild))
plt.plot(prix_a, label='Prix A', marker='o')
plt.plot(prix_b, label='Prix B', marker='x')
plt.axhline(y=20, color='red', linestyle='--', label='Benchmark')
plt.title("√âvolution des prix")
plt.xlabel("It√©ration")
plt.ylabel("Prix")
plt.legend()
plt.grid(True)
plt.show()


plt.plot(demandes_a, label='Demande A', marker='o')
plt.plot(demandes_b, label='Demande B', marker='x')
plt.title("√âvolution des demandes")
plt.xlabel("It√©ration")
plt.ylabel("Demandes")
plt.legend()
plt.grid(True)
plt.show()

plt.plot(profits_a, label='Profits A', marker='o')
plt.plot(profits_b, label='Profits B', marker='x')
plt.title("√âvolution des profits")
plt.xlabel("It√©ration")
plt.ylabel("profits")
plt.legend()
plt.grid(True)
plt.show()
```
